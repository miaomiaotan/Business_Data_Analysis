```python
import pandas as pd
import numpy as np
```

```python
sh = pd.read_csv('./Result/sh2016.csv')
```

```python
sh
```

# 0. relations

```python
relations = pd.pivot_table(sh, values='total_price', index='seller', columns='buyer', fill_value=0, aggfunc=sum)
```

```python
relations.to_csv('./Result/relations.csv')
```

# 1. trading by buyers

```python
buyers = sh.groupby('buyer',as_index=False).total_price.sum().sort_values('total_price', ascending = False)
```

```python
buyers['prop'] = buyers['total_price']/buyers.total_price.sum()
```

```python
buyers
```

```python
buyers.to_csv('./Result/buyers.csv')
```

# 2. top3 buyers

```python
buyer1 = sh[sh['buyer'] == '上海国家会计学院']
buyer2 = sh[sh['buyer'] == '上海出入境检验检疫局']
buyer3 = sh[sh['buyer'] == '同济大学']
```

```python
buyer1.to_csv('./Result/buyer1.csv')
buyer2.to_csv('./Result/buyer2.csv')
buyer3.to_csv('./Result/buyer3.csv')
```

# 3. trading by sellers

```python
sellers = sh.groupby('seller',as_index=False).total_price.sum().sort_values('total_price',ascending=False)
```

```python
sellers['prop'] = sellers['total_price']/sellers.total_price.sum()
```

```python
sellers
```

```python
sellers.to_csv('./Result/sellers.csv', encoding = 'gbk')
```
