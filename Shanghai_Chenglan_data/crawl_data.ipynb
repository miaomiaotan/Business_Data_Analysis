## a function: read url and return html

```python
def crawl_product_list(url) :
    
    null_proxy_handler = urllib2.ProxyHandler({}) 
    opener = urllib2.build_opener(null_proxy_handler)
    urllib2.install_opener(opener)
    request = urllib2.Request(url)

    user_agent = "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.1 (KHTML, like Gecko) Chrome/22.0.1207.1 Safari/537.1" 
    request.add_header('User-Agent', user_agent)
    try : 
        connection = opener.open(request, timeout=5)
        if connection.code == 200:
            html = connection.read()
            return html
        
    except urllib2.HTTPError, ex:
        print 'crawl_product_list() -------- ', str(ex)
        return None
    except urllib2.URLError, ex:
        print 'crawl_product_list() -------- ', str(ex)
        return None
    except Exception, ex:
        print 'crawl_product_list() -------- ', str(ex)
        return None
```

# write sales info to sale_info.csv

```python
with open('./Url/sales_information' , 'r') as r, open('./Data/sale_info.csv', 'w') as w:
    for line in r:
        sales_url = line
        try :
            raw_html = crawl_product_list(sales_url)
            soup = BeautifulSoup( raw_html , "lxml")
            #print sales_url           
            tds = soup.findAll("td", {"class" : "gys_style"})
        
            w.write( sales_url.split('=')[-1].strip() )
            for i in range( len(tds) ) :
                td_text = tds[i].renderContents()
                w.write( ' ' + td_text.strip() )
            w.write('\n')
        except Exception, ex:
            print 'spider_url_by_get（） -------- ', str(ex)
```

# write product parameter to product_type.csv

```python
with open('./Url/product_parameter' , 'r') as r, open('./Data/product_type.csv' , 'w') as w:
    for line in r:
        product_type = line
        print product_type
        try :
            raw_html = crawl_product_list(product_type)
            soup = BeautifulSoup( raw_html , "lxml")
            brand_model = soup.findAll('font',{'class':'Content'})

            w.write( product_type.split('=')[-1].strip() )
            for item in brand_model[0:2] :
                brand_info = item.text.encode('utf8')
                w.write(',' + brand_info)
            w.write('\n')

        except Exception, ex:
            print 'wirte file ex -------- ', str(ex)
```
