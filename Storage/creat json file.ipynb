```python
import pandas as pd
```

```python
models = pd.read_csv('./result/group_by_models.csv')
```

```python
models
```

```python
category = models['category'].tolist()
model = models['model'].tolist()
total_price = models['total_price'].tolist()
```

```python
with open('./Demo/models.json', 'w') as w:
    category.append(0)
    t = 0
    w.write('{\n "name": "server",\n "children": [\n')
    for i in range(len(category)-1):
        if category[i+1] != category[i]:
            w.write('  {\n   "name": "' +category[i]+'",\n   "children": [\n')
            for j in range(t,i+1):
                w.write('    {"name": "' + model[j] + '", "size": ' + str(total_price[j]) + '},\n') 
            w.write('   ]\n  },\n')
            t = i+1
    w.write(' ]\n}')
```
