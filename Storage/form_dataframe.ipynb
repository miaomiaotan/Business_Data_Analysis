```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1.read csv into dataframe

### 1.1 product_category

```python
col_names = ['id' , '首页' , '协议供货' , '信息类产品' , 'category']
```

```python
product_category = pd.read_csv('./Data/product_homepage.csv' , sep = ',', header = None , names = col_names , index_col = False)
```

```python
product_category.drop(product_category.columns[[1,2,3]], axis=1, inplace=True)
```

```python
product_category
```

### 1.2 product_type

```python
col_names = ['id' , 'brand' , 'type']
```

```python
product_type = pd.read_csv('./Data/product_parameter.csv' , sep = ',' , header = None , names = col_names , index_col = False)
```

```python
product_type
```

```python
product_type
```

### 1.3 product_sales

```python
col_names = ['id','buyer','number','price','seller','date','time']
```

```python
valid_sales = pd.read_csv('./Data/valid_sales.csv', sep = ' ', header = None, names = col_names, index_col = False )
```

```python
valid_sales
```

# 2.merge dataframe

```python
China = pd.merge(product_category , product_type ,  on = 'id' , how = 'inner').merge(valid_sales , on = 'id')
```

```python
China
```

```python
China['price'] = China['price'].str.replace(',','').astype(float)
```

```python
China['total_price'] = China['number'] * China['price']
```

```python
China
```

```python
China.to_csv('./Result/all_sale_info.csv', index=False)
```
