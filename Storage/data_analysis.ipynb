# 1.group by category

```python
categories = China.groupby('category', as_index=False).total_price.sum().sort_values('total_price',ascending=False)
```

```python
categories['prop'] = categories['total_price']/categories.total_price.sum()
```

```python
categories
```

```python
categories.to_csv('./Result/group_by_categories.csv' , index_col= None)
```

# 2.group by model

```python
def sort_models(group):
    return group.sort_values('total_price',ascending=False)
```

```python
models = China.groupby(['category','model'],as_index=False).total_price.sum()
```

```python
models['number'] = China.groupby(['category','model'],as_index=False).number.sum()['number']
```

```python
models = models.groupby(['category']).apply(sort_models)
```

```python
models['prop'] = models['total_price']/models.total_price.sum()
```

```python
models
```

```python
models.to_csv('./result/group_by_models.csv' , index_col= None)
```

# 2.5 models prop >= 0.015

```python
prop_0015 = models[models['prop'] >= 0.015]
```

```python
prop_0015
```

```python
prop_0015.to_csv('./result/prop>=0.015.csv' , index_col= None)
```

# 3.group by month

```python
def get_month(date):
    return date.split('-')[0] + '-' + date.split('-')[1]
```

```python
China['month'] = China['date'].apply(get_month)
```

```python
months = China.groupby('month', as_index=False).total_price.sum()
```

```python
months
```

```python
months.to_csv('./result/group_by_months.csv')
```
