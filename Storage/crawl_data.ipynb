# 1.creat search_list 

```python
with open('./storage.txt', 'r') as server, \
    open('./URL_pools/search_list.txt','w') as search_list:
    for line in server:
        for i in range(5):
            url = 'http://www.zycg.gov.cn/td_xxlcpxygh/itemsearch?KeyWord=' +str(line.strip())+ '&SearchType=Product&page=' + str(i+1)
            search_list.write(url + '\n')
```

# 2.visit urls in search_list, get html, use regex to get info

### a function: read url and return html

```python
def crawl_product_list(url) :
    
    null_proxy_handler = urllib2.ProxyHandler({}) 
    opener = urllib2.build_opener(null_proxy_handler)
    urllib2.install_opener(opener)
    request = urllib2.Request(url)

    user_agent = "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.1 (KHTML, like Gecko) Chrome/22.0.1207.1 Safari/537.1" 
    request.add_header('User-Agent', user_agent)
    try : 
        connection = opener.open(request, timeout=5)
        if connection.code == 200:
            html = connection.read()
            return html
        
    except urllib2.HTTPError, ex:
        print 'crawl_product_list() -------- ', str(ex)
        return None
    except urllib2.URLError, ex:
        print 'crawl_product_list() -------- ', str(ex)
        return None
    except Exception, ex:
        print 'crawl_product_list() -------- ', str(ex)
        return None
```

```python
with open('./URL_pools/search_list.txt', 'r') as search_list, \
open('./URL_pools/product_url', 'a') as product_url:
    product_pattern = re.compile('<a.*?>')
    for line in search_list:
        url = line
        result = crawl_product_list(url)
        try : 
            hrefs = re.findall(product_pattern, result)
            for href in hrefs:              
                if 'show_product' in href :
                    product_url.write(href + '\n')
        except Exception, ex:
            print 'spider_url_by_get（） -------- ', str(ex)
            continue
```

# 4.visit urls, write info to csv

### 4.1 product_category.csv

```python
with open('./URL_pools/product_homepage', 'r') as r,\
open('./Data/product_homepage.csv', 'w') as w:
    for line in r:
        sales_url = line
        print sales_url
        try :
            raw_html = crawl_product_list(sales_url)
            soup = BeautifulSoup( raw_html )
            category = soup.find_all("div",{"class":"location"})
            
            w.write( sales_url.split('/')[-1].strip() )
            for tag in category :
                aTags = tag.findAll('a')
                for aTag in aTags :
                    category_info = aTag.getText().encode('utf8')
                    w.write( ',' + category_info )
            w.write('\n')

        except Exception, ex:
            print 'wirte file ex -------- ', str(ex)
```

### 4.2 product_parameter.csv

```python
with open('./URL_pools/product_parameter', 'r') as r,\
open('./Data/product_parameter.csv', 'w') as w:
    for line in r:
        product_type = line
        print product_type
        try :
            raw_html = crawl_product_list(product_type)
            soup = BeautifulSoup( raw_html , "lxml")
            brand_model = soup.findAll('font',{'class':'Content'})

            w.write( product_type.split('=')[-1].strip() )
            for item in brand_model[0:2] :
                brand_info = item.text.encode('utf8').strip()
                w.write(',' + brand_info)
            w.write('\n')

        except Exception, ex:
            print 'wirte file ex -------- ', str(ex)
```

### 4.3 product_sales.csv

```python
with open('./URL_pools/product_sales' , 'r') as r, open('./Data/product_sales.csv', 'w') as w:
    for line in r:
        sales_url = line
        try :
            raw_html = crawl_product_list(sales_url)
            soup = BeautifulSoup( raw_html , "lxml")
            #print sales_url           
            tds = soup.findAll("td", {"class" : "gys_style"})
        
            w.write( sales_url.split('=')[-1].strip() )
            for i in range( len(tds) ) :
                td_text = tds[i].renderContents()
                w.write( ' ' + td_text.strip() )
            w.write('\n')
        except Exception, ex:
            print 'spider_url_by_get（） -------- ', str(ex)
```

```python
with open('./Data/product_sales.csv' , 'r') as r, \
open('./Data/valid_sales.csv' , 'w') as w:
    for line in r:
        line = line.strip()
        pro_id = line.split(' ')[0]
        info = line.split(' ')[7:]
        nrow = len(info)/7
        for row in range(nrow):
            w.write(pro_id + ' ')
            for i in range(7):
                if i != 0:
                    w.write(info[row*7+i] + ' ')
            w.write('\n')
```
